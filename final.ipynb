```python
water = dom['waterfront'].value_counts()
water.index = ['ничо не видно', 'водичка!']
plt.bar(water.index, water.values)
plt.xlabel('вид на набережную')
plt.ylabel('количество')
plt.title('вид на набережную');
```

```python
floor = dom['floors'].value_counts()
plt.bar(floor.index, floor.values, width=0.37)
plt.xlabel('количество этажей')
plt.ylabel('количество домов')
plt.title('этажность домов');
```

```python
cond = dom['condition'].value_counts().sort_index()
cond.index = ['хибары', 'ну такое', 'средне', 'годнота', 'ваще топ']
plt.bar(cond.index, cond.values, width=0.92)
plt.xlabel('степень пригодности')
plt.ylabel('количество домов')
plt.title('состояние');
```

...эммм чтож, данных, влияющих на цену не так и много, я возьму "класс" дома (grade), состояние, год постройки, количество спален и самое крутое - дата продажи!

```python
# №1
plt.scatter(dom['price'], dom['grade'])    # для начала возьму сравнение из самошо начала - по-моему оно самое показательное. Чем выше класс жилья - тем дороже есть дома, максимум 7.700.000
plt.xlabel('цена в млн')
plt.ylabel('класс дома')
plt.title('ценовой разброс по категориям');
```

```python
#   №2
plt.scatter(dom['price'], dom['condition'], s=3)    # А тут более странно. Самые классные дома, ну допустим, просто новые, не самые дорогие. А самые дорогие - какие-нибудь "поместья" в возрасте.
plt.xlabel('цена в млн')
plt.ylabel('состояние дома')
plt.yticks([1,2,3,4,5])  # я не придумал как округлить исходные значения
plt.title('ценовой разброс по состоянию');   # и всё-таки в "сплошной" массе синих точек неочевидно насколько их там много. Десятки и тысячи штук выглядят одинаково.
```

```python
#   №3
plt.scatter(dom['price'], dom['yr_built'], marker='.', c='brown', s=0.6)    # Ну чтож. Гипотеза про хотя бы одно "поместье" в возрасте подтвердилась :)))) 
plt.xlabel('цена в млн')                                                        # с таким размером точечек даже почти видно отдельные года. Ну и... серьёзного влияния на цену они не оказывают.
plt.ylabel('год постройки')
plt.title('ценовой разброс по возрасту');
```

```python
 # №3.1               А вот так чуть интереснее про года. Поскольку у нас ось времени = Y, то точка реновации всегда стоит НАД "своей" точкой постройки.
dom2 = dom['yr_renovated'].replace({0: np.NaN})     # сперва убираем нолики, а то нулевой год на графике не слишком способствует наглядности.
plt.scatter(dom['price'], dom['yr_built'], marker='.', c='black', s=3)   
plt.scatter(dom['price'], dom2, marker='.', c='red', s=1)     
plt.xlabel('цена в млн')                                                        
plt.ylabel('год постройки')
plt.title('ценовой разброс по возрасту');
```

```python
#   №4
plt.scatter(dom['price'], dom['bedrooms'], c='teal', s=3)    # Ну тут всё ясно. В самых дорогих домах 4-6 спален. А вот тот дворец с 33мя спальнями (есть там такой), он бюджетный. Может это коровник?
plt.xlabel('цена в млн')
plt.ylabel('количество спален')
plt.title('ценовой разброс по количеству спален');
```
